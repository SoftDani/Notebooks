# Precio articulos

```python
def precio_art (art,df):
    df_art=df[df['nomartic']==art]
    df_art_precio=df_art.groupby('anyparte').precio.mean()
    #display(df_ext_precio)
    df_art_preciouc=df_art.groupby('anyparte').preciouc.mean()
####
    print( 'Precio minimo ultima compra: ', round(df_art_preciouc.min(),2),'\n')
    print( 'Precio medio: ', round(df_art_precio.mean(),2),'\n')
###
    x_precio = df_art_precio.index
    y_precio= df_art_precio
    x_preciouc = df_art_preciouc.index
    y_preciouc=df_art_preciouc

    fig,axis = plt.subplots(1,1, figsize=(15,8))
    axis.plot(x_precio,y_precio,'go-',label='precio',linewidth=1,alpha=1) # creación del gráfico
    axis.plot(x_preciouc,y_preciouc,'bx-.',label='precio ultima compra',linewidth=1) # creación del gráfico
#
    axis.legend(loc='best')
    plt.title('precio frente a precio ultima compra')

    plt.grid()
    plt.show() 
```

```python
def gasto_bus(df, bus) :
    df_bus=df[df['codigbus']==bus]
    #df_bus.groupby('nomfamia').importe.max().sort_values(ascending=False)
    #df_bus.groupby('nomfamia').importe.max().sort_values(ascending=False)
    
    
    ### gasto Total
    x_bus=df_bus.groupby('nomfamia').importe.sum().sort_values(ascending=False)
    y_bus=df_bus.groupby('nomfamia').importe.sum().sort_values(ascending=False).index

    plt.figure(figsize=(15,15))
    sns.set(style='darkgrid')

    sns.barplot(x=x_bus, y=y_bus,data=df)
    plt.title('Autobus '+bus+' Gasto total por familia')
    plt.show()
  
    ##### Gasto total por año
    
    y_bus=df_bus.groupby('anyparte').importe.sum()
    x_bus=list(df_bus.groupby('anyparte').importe.sum().index)


    fig,axis = plt.subplots(1,1, figsize=(15,8))
    axis.plot(x_bus,y_bus,'go-',label='importe',linewidth=1,alpha=1) # creación del gráfico
    plt.xticks(x_bus)
    axis.legend(loc='best')
    plt.title('Autobus '+bus+' Gasto total por año')
    plt.grid()
    plt.show() 
    
   
    ##### detalles del gasto Total
    año_max = df_bus.groupby('anyparte').importe.sum().idxmax()
    print('\nAño con mayor gasto total, ', año_max ,'distribuido: \n')
    
    df_bus_año=df_bus[df_bus['anyparte']==año_max]
    gasto_año_max=df_bus_año.groupby('nomfamia').importe.sum().sort_values(ascending=False)
    display(gasto_año_max)
    
    print('\n')
    ### gasto medio
    x_bus=df_bus.groupby('nomfamia').importe.mean().sort_values(ascending=False)
    y_bus=df_bus.groupby('nomfamia').importe.mean().sort_values(ascending=False).index

    plt.figure(figsize=(15,15))
    sns.set(style='darkgrid')

    sns.barplot(x=x_bus, y=y_bus,data=df)
    plt.title('Autobus '+bus+' Gasto medio por familia')
    plt.show()
    #####
    ##### Gasto medio por año
    
    y_bus=df_bus.groupby('anyparte').importe.mean()
    x_bus=list(df_bus.groupby('anyparte').importe.mean().index)


    fig,axis = plt.subplots(1,1, figsize=(15,8))
    axis.plot(x_bus,y_bus,'go-',label='importe',linewidth=1,alpha=1) # creación del gráfico
    plt.xticks(x_bus)
    axis.legend(loc='best')
    plt.title('Autobus '+bus+' Gasto  medio por año')
    plt.grid()
    plt.show() 
    
    ##### detalles del gasto medio
    año_max = df_bus.groupby('anyparte').importe.mean().idxmax()
    print('\nAño con mayor gasto medio, ', año_max ,'distribuido: \n')
    
    df_bus_año=df_bus[df_bus['anyparte']==año_max]
    gasto_año_max=df_bus_año.groupby('nomfamia').importe.mean().sort_values(ascending=False)
    display(gasto_año_max)
    
    #display(df_bus.sample(5))
    
    
```

# Gasto por Empresa

```python
def gasto_empresa(df, emp) :
    df_emp=df[df['nomempre']==emp]
    
    
    
    #Gasto total por familia
    x_emp=df_emp.groupby('nomfamia').importe.sum().sort_values(ascending=False)
    y_emp=df_emp.groupby('nomfamia').importe.sum().sort_values(ascending=False).index

    plt.figure(figsize=(15,15))
    sns.set(style='darkgrid')

    sns.barplot(x=x_emp, y=y_emp,data=df)
    plt.title('Empresa '+emp+' Gasto total por familia')
    plt.show()
    

    
    ##### Gasto total por año
    
    y_emp=df_emp.groupby('anyparte').importe.sum()
    x_emp=list(df_emp.groupby('anyparte').importe.sum().index)


    fig,axis = plt.subplots(1,1, figsize=(15,8))
    axis.plot(x_emp,y_emp,'go-',label='importe',linewidth=1,alpha=1) # creación del gráfico
    plt.xticks(x_emp)
    axis.legend(loc='best')
    plt.title('Empresa '+emp+' Gasto total por año')
    plt.grid()
    plt.show() 
    
   
    #####
    año_max = df_emp.groupby('anyparte').importe.sum().idxmax()
    print(f'\nAño con mayor gasto total en {emp} es, {año_max} distribuido: \n')
    
    df_emp_año=df_emp[df_emp['anyparte']==año_max]
    gasto_año_max=df_emp_año.groupby('nomfamia').importe.sum().sort_values(ascending=False)
    display(gasto_año_max)
    
    #####
    
    print('\n')
    ### gasto medio
    x_emp=df_emp.groupby('nomfamia').importe.mean().sort_values(ascending=False)
    y_emp=df_emp.groupby('nomfamia').importe.mean().sort_values(ascending=False).index

    plt.figure(figsize=(15,15))
    sns.set(style='darkgrid')

    sns.barplot(x=x_emp, y=y_emp,data=df)
    plt.title('Empresa '+emp+' Gasto medio por familia')
    plt.show()
    #####
    
    ##### Gasto medio por año
    
    y_emp=df_emp.groupby('anyparte').importe.mean()
    x_emp=list(df_emp.groupby('anyparte').importe.mean().index)


    fig,axis = plt.subplots(1,1, figsize=(15,8))
    axis.plot(x_emp,y_emp,'go-',label='importe',linewidth=1,alpha=1) # creación del gráfico
    plt.xticks(x_emp)
    axis.legend(loc='best')
    plt.title('Empresa '+emp+' Gasto  medio por año')
    plt.grid()
    plt.show() 
    
    ##### detalles del gasto medio
    año_max = df_emp.groupby('anyparte').importe.mean().idxmax()
    print(f'\nAño con mayor gasto medio en {emp} es, {año_max} distribuido: \n')
    
    df_emp_año=df_emp[df_emp['anyparte']==año_max]
    gasto_año_max=df_emp_año.groupby('nomfamia').importe.mean().sort_values(ascending=False)
    display(gasto_año_max)
    
    #display(df_bus.sample(5))
    
```
