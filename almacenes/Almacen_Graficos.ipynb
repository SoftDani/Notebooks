# Precio Extintores

```python
x = df_ext_precio.index
y_precio= df_ext_precio
y_preciouc=df_ext_preciouc

fig,axis = plt.subplots(1,1, figsize=(15,8))
axis.plot(x,y_precio,'go-',label='precio',linewidth=1,alpha=1) # creación del gráfico
axis.plot(x,y_preciouc,'bx-.',label='precio ultima compra',linewidth=1) # creación del gráfico
#
axis.legend(loc='best')
plt.grid()
plt.show() 
```
